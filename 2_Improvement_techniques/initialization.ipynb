```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Dropout
from keras import Sequential
```

```python
from keras.initializers import Zeros,RandomUniform,RandomNormal,GlorotNormal,GlorotUniform,HeNormal,HeUniform
```

```python
cancer = load_breast_cancer()

df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X=cancer.data
y=cancer.target

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
model = Sequential([
    Dense(128, input_dim=30, activation='relu', kernel_initializer=Zeros()),
    Dense(64, activation='relu', kernel_initializer=Zeros()),
    Dense(8, activation='relu', kernel_initializer=Zeros()),
    Dense(1, activation='sigmoid', kernel_initializer=Zeros()),
])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

Randomuniform

```python
model = Sequential([
    Dense(128, input_dim=30, activation='relu', kernel_initializer=RandomUniform(minval=-0.05, maxval=0.05)),
    Dense(64, activation='relu', kernel_initializer=RandomUniform(minval=-0.05, maxval=0.05)),
    Dense(8, activation='relu', kernel_initializer=RandomUniform(minval=-0.05, maxval=0.05)),
    Dense(1, activation='sigmoid', kernel_initializer=RandomUniform(minval=-0.05, maxval=0.05)),
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=25, validation_split=0.2)
```
