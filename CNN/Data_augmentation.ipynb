```python
from keras.preprocessing import image
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/cat-with-open-mouth.webp',target_size=(256,256))
```

```python
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=25,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode="constant"
)
```

```python
img_arr=image.img_to_array(img)
```

```python
type(img_arr)
```

```python
input_img=img_arr.reshape(1,256,256,3)
```

```python
i=0
for output in datagen.flow(input_img,batch_size=1,save_to_dir='/content/aug'):
  i=i+1
  if i==10:
    break
```
